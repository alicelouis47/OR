ตัวอย่างที่ 12.1: การเสียของเครื่องจักร
โจทย์: จำลองเหตุการณ์ 365 วัน โดยมีโอกาสที่เครื่องจักรจะเสีย 1 ใน 6 (ความน่าจะเป็น 0.167) ถ้าเสียจะมีค่าใช้จ่าย 1,000 บาท 


โค้ดนี้จะจำลองการ "ทอยลูกเต๋า" 365 ครั้ง ถ้าได้เลข 1 ถือว่าเครื่องเสีย

```python
import random

# --- การตั้งค่าจากตัวอย่าง 12.1 ---
N_DAYS = 365
COST_PER_FAILURE = 1000
# ความน่าจะเป็นที่เครื่องจะเสียคือ 1 ใน 6 (เหมือนทอยลูกเต๋าได้เลข 1) [cite: 7, 10]

total_failures = 0

# --- ทำการจำลองเหตุการณ์ 365 วัน (365 ครั้ง) --- 
for day in range(N_DAYS):
    # สุ่มทอยลูกเต๋า 1 ลูก (เลข 1 ถึง 6)
    die_roll = random.randint(1, 6)
    
    # ถ้าได้แต้ม 1 ถือว่าเครื่องเสีย 
    if die_roll == 1:
        total_failures += 1
        
# --- สรุปผล ---
total_cost = total_failures * COST_PER_FAILURE

print(f"--- ผลการจำลองตัวอย่างที่ 12.1 (365 วัน) ---")
print(f"จำนวนวันที่เครื่องจักรเสีย (จำลอง): {total_failures} วัน")
print(f"ค่าใช้จ่ายโดยประมาณ: {total_cost:,.0f} บาท")

# เปรียบเทียบกับค่าทางทฤษฎี (Expected Value)
expected_failures = N_DAYS * (1/6)
expected_cost = expected_failures * COST_PER_FAILURE
print(f"ค่าคาดหวัง (ทฤษฎี): {expected_failures:.2f} วัน (ค่าใช้จ่าย {expected_cost:,.0f} บาท)")
print(f"*(ผลจากการรันโค้ดแต่ละครั้งจะไม่เท่ากัน แต่จะใกล้เคียงกับค่าคาดหวัง [cite: 11])")
```

ตัวอย่างที่ 12.2: การตัดสินใจสั่งสินค้า
โจทย์: จำลองเหตุการณ์ 50 วัน เพื่อเปรียบเทียบกำไรระหว่างการสั่งสินค้า 5, 6, และ 7 หน่วย 



ข้อมูล: ต้นทุน 12 บาท, ราคาขาย 15 บาท (กำไร 3 บาท), ขายไม่ออกเหลือ 6 บาท (ขาดทุน 6 บาท) 


การจำลอง: ใช้การสุ่มเลข 00-99 เพื่อจำลองความต้องการ (Demand) ในแต่ละวันตามตารางความน่าจะเป็น

```python
import random

# --- การตั้งค่าจากตัวอย่าง 12.2 ---
N_DAYS = 50
COST_PRICE = 12         # 
SELL_PRICE = 15         # 
SALVAGE_PRICE = 6       # 

PROFIT_PER_UNIT = SELL_PRICE - COST_PRICE
LOSS_PER_UNSOLD = COST_PRICE - SALVAGE_PRICE

# [อ้างอิงจากตารางหน้า 156 (source 23)]
def get_simulated_demand():
    """
    จำลองความต้องการสินค้าโดยใช้เลขสุ่ม 2 หลัก (00-99)
    ตามช่วงความน่าจะเป็นสะสมในตาราง 
    """
    r = random.randint(0, 99) # สุ่มเลข 00-99
    if 0 <= r <= 9:     # 00-09
        return 3
    elif 10 <= r <= 29:   # 10-29
        return 4
    elif 30 <= r <= 59:   # 30-59
        return 5
    elif 60 <= r <= 79:   # 60-79
        return 6
    elif 80 <= r <= 89:   # 80-89
        return 7
    elif 90 <= r <= 99:   # 90-99
        return 8

def calculate_profit(order_qty, demand):
    """
    คำนวณกำไร/ขาดทุนตามตรรกะของ Pay off Table [cite: 38]
    """
    if demand >= order_qty:
        # ถ้าความต้องการ >= ที่สั่ง (ขายหมด)
        profit = order_qty * PROFIT_PER_UNIT
    else:
        # ถ้าความต้องการ < ที่สั่ง (ขายไม่หมด)
        units_sold = demand
        units_unsold = order_qty - demand
        profit = (units_sold * PROFIT_PER_UNIT) - (units_unsold * LOSS_PER_UNSOLD)
    return profit

# --- ทำการจำลองเหตุการณ์ 50 วัน --- 
order_quantities_to_test = [5, 6, 7] # 
total_profits = {qty: 0 for qty in order_quantities_to_test}
demand_counts = {d: 0 for d in [3,4,5,6,7,8]} # สำหรับนับความถี่

for day in range(1, N_DAYS + 1):
    demand = get_simulated_demand()
    demand_counts[demand] += 1
    
    # คำนวณกำไรสำหรับแต่ละทางเลือก
    for order_qty in order_quantities_to_test:
        profit_today = calculate_profit(order_qty, demand)
        total_profits[order_qty] += profit_today

# --- สรุปผล ---
print(f"\n--- ผลการจำลองตัวอย่างที่ 12.2 (50 วัน) ---")
print("*(ตัวเลขกำไรจะไม่ตรงกับในหนังสือ [cite: 43] เพราะใช้เลขสุ่มคนละชุด)")
print("เปรียบเทียบกำไรรวมจากการสั่งซื้อ:")
best_qty = 0
max_profit = -float('inf')

for qty, total_profit in total_profits.items():
    avg_profit = total_profit / N_DAYS
    print(f"  - สั่ง {qty} หน่วย: กำไรรวม = {total_profit:,.0f} บาท (เฉลี่ย {avg_profit:,.2f} บาท/วัน)")
    if total_profit > max_profit:
        max_profit = total_profit
        best_qty = qty

print(f"\nผลลัพธ์นี้สนับสนุนการตัดสินใจสั่งสินค้า {best_qty} หน่วย ซึ่งให้กำไรสูงสุด [cite: 45]")

# เปรียบเทียบความถี่ที่จำลองได้กับค่าสถิติเดิม (เหมือนตารางหน้า 158 [cite: 34])
print("\n--- เปรียบเทียบความถี่ความต้องการ (จำลอง vs สถิติ) ---")
print(f"{'ความต้องการ':<10} {'จำลอง (ครั้ง)':<12} {'จำลอง (%)':<10} {'สถิติ (%)':<10}")
stats_prob = {3: 0.1, 4: 0.2, 5: 0.3, 6: 0.2, 7: 0.1, 8: 0.1} # [cite: 18]
for demand, count in demand_counts.items():
    prob_sim = (count / N_DAYS) * 100
    prob_stat = stats_prob[demand] * 100
    print(f"{demand:<10} {count:<12} {prob_sim:<10.1f} {prob_stat:<10.1f}")
```

ตัวอย่างที่ 12.3: ตัวแบบแถวคอย (Queuing Model)
โจทย์: จำลองเหตุการณ์ 15 นาที เพื่อเปรียบเทียบค่าใช้จ่ายรวม (ค่าจ้างพนักงาน + ค่าเสียโอกาส) ระหว่างการจ้างพนักงาน 1, 2, 3, และ 4 คน 
ข้อมูล:ลูกค้ามาแบบสุ่ม (Poisson $\lambda=3$)(ใช้ตารางสุ่ม 000-999) 
ระบบรับได้สูงสุด 6 คน ถ้าเต็ม ลูกค้าใหม่จะออกไป (OUT) ค่าเสียโอกาสลูกค้าออกไป (OUT) คนละ 20 บาท 15ค่าจ้างพนักงานคนละ 100 บาท 
พนักงาน 1 คน ให้บริการได้ 1 คน/นาที (Service rate = 1) 

```python
import random

# --- การตั้งค่าจากตัวอย่าง 12.3 ---
N_MINUTES = 15            # 
MAX_IN_SYSTEM = 6       # (รองรับได้สูงสุด 6 คน) 
COST_LOST_CUSTOMER = 20   # 
COST_PER_SERVER = 100     # 
SERVICE_RATE_PER_SERVER = 1 # (ให้บริการได้ 1 คน/นาที) 

# [อ้างอิงจากตารางหน้า 162 (source 69)]
def get_simulated_arrivals():
    """
    จำลองจำนวนลูกค้าที่เข้ามาใน 1 นาที โดยใช้เลขสุ่ม 3 หลัก (000-999)
    ตามช่วงความน่าจะเป็นสะสมของ Poisson (mean=3) 
    """
    r = random.randint(0, 999) # สุ่มเลข 000-999
    if 0 <= r <= 49:     # 000-049
        return 0
    elif 50 <= r <= 198:   # 050-198
        return 1
    elif 199 <= r <= 422:  # 199-422
        return 2
    elif 423 <= r <= 626:  # 423-626
        return 3
    elif 627 <= r <= 814:  # 627-814
        return 4
    elif 815 <= r <= 915:  # 815-915
        return 5
    else: # 916-999
        return 6 # (ใน text คือ 6 และ มากกว่า) 

server_options = [1, 2, 3, 4] # 
final_results = {}

print(f"--- ผลการจำลองตัวอย่างที่ 12.3 (15 นาที) ---")
print("*(ตัวเลขจะไม่ตรงกับตารางในหนังสือ [cite: 81] เพราะใช้เลขสุ่มคนละชุด)")

# --- ทำการจำลองเหตุการณ์ 15 นาที สำหรับพนักงานแต่ละจำนวน ---
for n_servers in server_options:
    
    total_lost_customers = 0
    # Lq คือ จำนวนลูกค้าที่อยู่ในระบบหลังให้บริการ (ณ สิ้นนาที) [cite: 77]
    lq_at_end_of_minute = 0 
    
    for minute in range(1, N_MINUTES + 1):
        lq_start = lq_at_end_of_minute
        
        # 1. ลูกค้าใหม่เข้ามา
        arrivals = get_simulated_arrivals()
        
        # 2. คำนวณคนเข้าระบบ (IN) และคนไม่เข้า (OUT) [cite: 73, 76]
        system_before_service = lq_start + arrivals
        
        if system_before_service > MAX_IN_SYSTEM:
            customers_lost = system_before_service - MAX_IN_SYSTEM
            customers_in = MAX_IN_SYSTEM - lq_start # คนที่เข้ามาได้จริง
        else:
            customers_lost = 0
            customers_in = arrivals
            
        total_lost_customers += customers_lost
        
        # 3. คำนวณการให้บริการ
        system_after_arrivals = lq_start + customers_in
        # พนักงาน n_servers คน บริการได้ n_servers * 1 คน/นาที
        served = min(system_after_arrivals, n_servers * SERVICE_RATE_PER_SERVER)
        
        # 4. อัปเดต Lq สิ้นนาที
        lq_at_end_of_minute = system_after_arrivals - served
        
    # --- สรุปค่าใช้จ่ายสำหรับ n_servers นี้ ---
    server_cost = n_servers * COST_PER_SERVER           # [cite: 87, 88, 89, 92]
    lost_customer_cost = total_lost_customers * COST_LOST_CUSTOMER # [cite: 82, 83, 84, 85]
    total_cost = server_cost + lost_customer_cost     # [cite: 86]
    
    final_results[n_servers] = {
        'total_lost': total_lost_customers,
        'lost_cost': lost_customer_cost,
        'server_cost': server_cost,
        'total_cost': total_cost
    }

# --- พิมพ์ตารางสรุปผลเปรียบเทียบ ---
print("\n--- ตารางสรุปค่าใช้จ่ายรวม ---")
print(f"{'จำนวนพนักงาน':<12} {'ลูกค้าไม่เข้า (OUT)':<15} {'ค่าเสียโอกาส (บาท)':<16} {'ค่าจ้าง (บาท)':<12} {'ค่าใช้จ่ายรวม (บาท)':<17}")

min_cost = float('inf')
best_option = 0

for n_servers, data in final_results.items():
    print(f"{n_servers:<12} {data['total_lost']:<15} {data['lost_cost']:<16,.0f} {data['server_cost']:<12,.0f} {data['total_cost']:<17,.0f}")
    if data['total_cost'] < min_cost:
        min_cost = data['total_cost']
        best_option = n_servers
        
print(f"\nผลการจำลองสรุปว่า ควรจ้างพนักงาน {best_option} คน (มีค่าใช้จ่ายรวมต่ำที่สุด) [cite: 94]")
```
